# Decision Tree Clasifier

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Day7.csv')
#dataset
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
#Independent variable
X = dataset.iloc[:,:-1].values

#Dependent variable
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#Model building with Decision Tree
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier.fit(X_train,y_train)
```

```python
#Prediction
classifier.predict(sc.transform([[40,98000]]))
```

```python
from sklearn import tree
fig = plt.figure(figsize=(20,20))
_=tree.plot_tree(classifier,feature_names=dataset.columns,class_names='Purchased',filled=True)
```

# Random Forest Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Day7.csv')
#dataset
```

```python
#Features
X = dataset.iloc[:,:-1].values

#Classifier
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier1 = RandomForestClassifier(n_estimators=200, criterion='entropy',random_state=0)
classifier1.fit(X_train,y_train)
```

```python
#Prediction
classifier1.predict(sc.transform([[40,98000]]))
```
