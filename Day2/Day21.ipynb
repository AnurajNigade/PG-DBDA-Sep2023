# Matplotlib Library

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Ex: 2-D Scatter plot
x=np.arange(0,10)
y=np.arange(10,20)
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Testing data plot')
```

```python
#Ex: 2-D plot
y=x*x
plt.plot(x,y,'g*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Testing data plot')
```

```python
#Ex:Subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*--')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')

plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
#Ex: 
x=np.arange(1,11)
y = 3*x+5
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Testing data plot')
```
